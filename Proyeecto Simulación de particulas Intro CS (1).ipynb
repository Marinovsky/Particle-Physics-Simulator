#Caída Libre y rebote de particulas
Para simular la caída libre de un objeto, se ha utilizado el Principio fundamental de la dinámica propuesto por Isaac Newton en 1687, el cual describe la relación entre la fuerza, masa y aceleración de un objeto, y permite hallar los valores de la aceleración, la velocidad y la posición, tan solo a partir de fuerza y de la masa. Además, también se hizo un analisis del problema como si fuera tiro parabólico, el cual nos ayuda a modelar el comportamiento de la caida de un partícula con una fuerza inicial en el eje *y* nula.

La principal formula incorporadas del principio fundamental de la dinámica es la que se  presenta a continuación:
* $F_{resultante}=m*a$

En esta formula la fuerza() es directamente proporcional al producto entre la masa(m) y la aceleración(a). Es justo a partir de está formula dónde con el uso de algebra podemos a partir de la fuerza y la masa hallar la aceleración:
* $a={\frac{F}{m}}$

Otro punto importante que es necesario aclarar, es que al estar hablando de caida libre y el rebote de una pariticula, la fuerza total  se va a descomponer en la fuerza en el eje y en el eje x. La fuerza en el eje y está dada especialemente por la fuerza de la gravedad, ya que está fuerza es en si la que hace que el objeto caiga. Por otro lado,  la fuerza en el eje x esta dada por la fuerza que nosotros le demos menos la resistencia del aire.

*  $Fy=Fy+gravity$ 
* $Fx=Fx-air_{resistence}$

Como la aceleración es un cambio en la velocidad, podemos definir la velocidad en terminos de esta, con la siguiente formula diferencial $dx=dx+ax$ dónde $dx$ esta la velocidad en el eje x, y $ax$ es el cambio que sufrió, es decir, la aceleración en eje x. De igual modo la velocidad en el eje y, está dada por la siguiente formula $dy=dy+ay$ dónde $dy$ es la velocidad en el eje y, y $ay$ la aceleración en el eje y.

De igual modo a partir de la velocidad podemos determinar la posición, puesto que la velocidad es un cambio de la posición en determinado tiempo. Así la formula diferencial que utilizamos es la que se presenta a continuación:

* $xv = xv + dx$
* $yv = yv + dy$

En las anteriores ecuaciones tanto *xv* como *yv* son las respectivas posiciones de la paritcula en  los ejes *x* e *y*, Y *dy* y *dx* son las velocidades en los ejes y, x.

Un importante detalle que cabe aclarar, es que para simular el robote de la particula hemos puesto una variable de tipo booleano llamada $condition$ con la cual cambia su estado de $False$ a $True$ cuando la posición de la particula es menor o igual a 0, que es el supuesto "piso" que hemos implementado por cuestiones estéticas; Esto hace que se cambie la dirección de la particula y que se vaya disminuyendo la energía de esta, que es la que hace que siga rebotando, hasta que pierda toda su energía.

#Caida en sustancias viscosas

Para simular este movimiento de la partícula, además del ya mencionado Principio fundamental de la dinámica, se ha utlizado la Ley de Stokes, derivada por el irlandés George Gabriel Stokes en 1851, en la cual se describe el comportamiento de un objeto inmerso en un fluido viscoso.

La Ley de Stokes describe la fuerza de fricción experimentada por objetos esféricos moviéndose en el seno de un fluido viscoso en un régimen laminar de bajos números de Reynolds. Se llama flujo laminar o corriente laminar, al movimiento de un fluido cuando éste es ordenado, estratificado, suave. 

* $F_d=6\pi R \eta v$

En la anterior formula $R$ es el radio de la esfera, $v$ su velocidad y $η$ la viscosidad del fluido.

En general la ley de Stokes es válida en el movimiento de partículas esféricas pequeñas moviéndose a velocidades bajas.
 
Puesto que antes de caer, la particula se encuentra en el aire, la fuerza está dada por el principio fundamental de la dinámica el cual mencionamos en la *Caida libre*, pero luego, cuando la partícula entra en el fluido viscoso, dónde ahora la fuerza está definida por la Ley de Stokes. Ante está problemática definimos en el código una variable llamada *state* la cual en principio es igual a *freefall*, pero cuando entra en el medio viscoso cambia su valor a *viscosity*, de esta manera podemos implementar los diferentes comportamientos que tiene la particula en la caida al medio viscoso.

# Resortes y Muelles

Con el fin de simular el movimiento de un resorte, se utlizó la ley de elasticidad de Hooke, propuesta en el año 1660 por el físico inglés Robert Hooke. Esta ley establece que el alargamiento de un muelle es directamente proporcional al módulo de la fuerza que sea aplicado sobre dicho muelle, siempre y cuando no se mantenga deformado el muelle. Por lo general la forma mas común de representar la matemáticamente dicha ley, es mediante la ecuación del resorte o del muelle, en la cual se relaciona la fuerza $F$ ejercida por el resorte, con la elongación $\delta$ dada por la fuerza externa aplicada al extremo del mismo, y sin olvidar la constante elástica del resorte denotada por la letra $k$.
* ![texto alternativo](https://wikimedia.org/api/rest_v1/media/math/render/svg/7215d534022abea68292b2aa0140f232ca1f8358)

Para simular el comportamiento del resorte no fue necesario en si comenzar de ceros, ya que las mismas formulas diferenciales utilizadas para calcular la aceleración, la velocidad y la posición fueron reutilizadas en este código, ya que todas ellas se puede calcular con la fuerza y la masa de la particula. Un importante detalle que cabe aclarar, es que hemos añadido una variable $dt$ la cual incluimos en el cálculo de la velocidad y la posición, esta variable $dt$ nos ayuda a estabilizar la velocidad de la particula. No obstante también se la puede eliminar y disminuir el valor de la constante $k$, para dismunuir la velocidad de la particula.

# Caída libre y rebote:










```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

xv = 0;
yv = 1000;

x=[]
y=[]

dx = 0;
dy = 0;
w = 0
condition=False

ax=0
ay=0

Fx=10
Fy=0
gravity=9.8
air_force=0.5
m=1


fig, ax = plt.subplots()
plt.close()
ax.set_xlim(( -10000, 10000))
ax.set_ylim((-2000, 2000))
line2, = ax.plot([], [], 'ro')
plt.setp(line2, markersize=10)
line1, = ax.plot([], [], lw=2) #The floor

def init():
    for i in range(-10000,10000):
      x.append(i)
      y.append(-50)
    line1.set_data(x, y)      
    return (line1,)

# animation function: this is called sequentially


def animate(i):

  global xv;
  global yv;
  global Fy;
  global Fx;
  global gravity;
  global air_force;
  global dx;
  global dy;
  global ax;
  global ay;
  global m;
  global w;
  global condition;

  # Configuramos las fuerzas
  Fy=Fy+gravity
  Fx=Fx-air_force
  
  # Configuramos las aceleraciones
  ax  = Fx/m
  ay  = Fy/m
  
  # Actualizamos las velocidades
  
    # En el eje x
  dx = dx + ax;
  
    #Este if es para asegurarnos que la velocidad no se vuelva negativa
  
  if dx<=0:
    dx=0
  
    # En el eje y
  if condition==False:
    dy = dy - ay;
  else:
    dy = 0
    
  # actualizar posiciones
  
    # En el eje x
  if dx==0:
    xv=xv
  else:
    xv = xv + dx
    
    # En el eje y
  if condition==False: #Para que no siga utlizando las mismas formulas después del primer robote de la partícula
    if yv+dy<=0:
      yv=0
    else:
      yv = yv + dy
  else:
    yv=0
  if yv<=0:
    if dy<=0:
      dy=dy*-1
      if (yv+dy)<=0:
        condition=True
    else:
      condition=True
      

  line2.set_data([xv],[yv])
  return (line2,)
  
anim = animation.FuncAnimation(fig, animate,init_func=init, frames=100, interval=50, blit=True)
rc('animation', html='jshtml')
anim
```

# Caída en sustancias viscosas

```
import matplotlib.pyplot as plt
from matplotlib import animation, rc

fig, ax = plt.subplots()

ax.set_xlim(( -100, 50000))
ax.set_ylim(( -20000, 20000))
line1, = ax.plot([], [], 'ro')
plt.setp(line1, markersize=10)
line2, = ax.plot([], [], lw=5)

y1=[]
x1=[]
x=0;
y=10000;
dx = 0;
dy = 0;

ax=0
ay=0

Fx=-10
Fy=-10
gravity=0.8
air_force=0.02
m=1
state="freefall"
coef_viscosity=0.01


def init():
    for i in range(0,50000):
      x1.append(i)
      y1.append(0)
    line2.set_data(x1, y1)      
    return (line2,)

def animate(i):
  global Fx;
  global Fy;
  global m;
  global gravity;
  global air_force
  global dy;
  global dx;
  global ax;
  global ay;
  global x;
  global y;
  global state;
  global coef_viscosity;
  
  if state=="freefall":
    #Comportamiento de caída libre
    Fy=Fy+gravity
    Fx=Fx+air_force
  elif state=="viscosity":
    #Comportamiento de la partícula en una sustancia viscosa
    Fy=6.0 * coef_viscosity * dy
    Fx=6 *coef_viscosity * dx
    
  # Configuramos las aceleraciones
  
  ax  = Fx/m
  ay  = Fy/m
  
  # actualizamos las velocidades
  dx = dx - ax;
  dy = dy - ay;
  
  # actualizamos las posiciones
  x=x+dx
  y=y+dy
  if y<=0 and state =="freefall": #Cuando la posición de la partícula es menor a cero, signifpica que ya entró en el medio viscoso
    state="viscosity"
    dx = dx/1.2
    dy = dy/1.2
    #Esta actualización de las velocidades es porque al caer a perdido velocidad
    
  line1.set_data([x],[y]);
  
  return (line1,)
  

anim = animation.FuncAnimation(fig, animate, init_func = init,frames=200, interval=10, blit=True)
rc('animation', html='jshtml')
anim
```

# Resortes y Muelles

```
import matplotlib.pyplot as plt
from matplotlib import animation, rc

fig, ax = plt.subplots()

ax.set_xlim(( -10, 10))
ax.set_ylim(( -10, 20))
line1, = ax.plot([], [], 'ro')
plt.setp(line1, markersize=10)
line2, = ax.plot([], [], lw=1)

y1=[]
x1=[]
x=0;
y=0;
vx = 0;
vy = 0;
ax = 0;
ay = 0;
dt = 0.1;
k = 1;


def init():
    for i in range(0,2000):
      x1.append(0)
      y1.append(i)
    line2.set_data(x1, y1)      
    return (line2,)

def animate(i):
  
  global vy;
  global vx;
  global ax;
  global ay;
  global x;
  global y;
  global k;
  
  #Configuramos las fuerzas
  Fy =10 - k*abs(y);
  
  #Configuramos las aceleraciones
  ay = Fy;
  
  # Actualizamos las velocidades
  vy = vy +  dt*ay;
  vx = vx +  dt*ax;
  
  # Actualizamos las posiciones
  x = x +   dt*vx;
  y = y +   dt*vy;
  
  line1.set_data([x],[y]);
  
  return (line1,)

anim = animation.FuncAnimation(fig, animate, init_func = init,frames=500, interval=10, blit=True)
rc('animation', html='jshtml')
anim
```
